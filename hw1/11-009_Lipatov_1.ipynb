```python
from matplotlib.image import imread, imsave
import matplotlib.pyplot as plt
import numpy as np

orig_img = imread('01.jpg')


def censor(orig, y_min, y_max, x_min, x_max, patches_count=10):
    img = np.asarray(orig.copy())

    patch_height = (y_max - y_min) // patches_count
    patch_width = (x_max - x_min) // patches_count

    for x in range(patches_count):
        for y in range(patches_count):
            y0 = y_min + patch_height * y
            y1 = y0 + patch_height
            x0 = x_min + patch_width * x
            x1 = x0 + patch_width

            img[y0:y1, x0:x1] = np.mean(img[y0:y1, x0:x1], axis=(0, 1))

    return img

censored = censor(orig_img, 450, 650, 400, 600, 5)
plt.imshow(censored)
plt.show()

imsave('01_censored.jpg', censored)
```
